```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/merged_2016.csv')
df.head()
```

```python
# Encode object type columns to numerical values
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for c in df.columns:
    df[c] = df[c].fillna(0)
    if df[c].dtype == 'object':
        le.fit(list(df[c].values))
        df[c] = le.transform(list(df[c].values))
        
df.columns
```
